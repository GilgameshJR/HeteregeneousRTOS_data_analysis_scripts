```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as tkr
```

```python
fft_df=pd.read_csv("fft_8_fn_vs_threshold.csv", delimiter=";")
```

```python
fft_df.head()
```

```python
fft_df.dropna(axis=1, inplace=True)
```

```python
ann_df=pd.read_csv("ann_fn_vs_threshold.csv", delimiter=";")
```

```python
ann_df.dropna(axis=1, inplace=True)
```

```python
gauss_df=pd.read_csv("gauss_fn_vs_threshold.csv", delimiter=";")
```

```python
gauss_df.dropna(axis=1, inplace=True)
```

```python
latnav_df=pd.read_csv("latnav_2_fn_vs_threshold.csv", delimiter=";")
```

```python
latnav_df.dropna(axis=1, inplace=True)
```

```python
fig = plt.plot(latnav_df["threshold"], latnav_df["fn_rate"], label="Latnav")
plt.plot(ann_df["threshold"], ann_df["fn_rate"], label="ANN")
plt.plot(gauss_df["threshold"], gauss_df["fn_rate"], label="Gaussian blur")
plt.plot(fft_df["threshold"], fft_df["fn_rate"], label="FFT")

plt.grid(visible=True, which='major', color='0.6')
plt.grid(visible=True, which='minor', color='0.8')

plt.yscale('log')
plt.xscale('log')
ax=plt.gca()
# #ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x,pos: ('{:d}%'.format(int(x)))))

plt.tick_params(axis='y', which='minor')
plt.tick_params(axis='x', which='minor')


plt.xlabel("Accepted relative error threshold [%]")
plt.ylabel("False negatives rate [%]")

plt.xlim((0.01, 1000))
plt.ylim((0.001, 100))

plt.legend()
```
